```python
#Calculs 
list_ind_score=['score_moral','score_terrain','score_teletrav','score_chomag','score_engagement','score_manag']
list_ind=['ind_moral','ind_terrain','ind_teletrav','ind_chomag','ind_engagement','ind_manag']

data_analyse['score_moral']=''
data_analyse['score_terrain']=''
data_analyse['score_teletrav']=''
data_analyse['score_chomag']=''
data_analyse['score_engagement']=''
data_analyse['score_manag']=''

data_analyse['ind_moral']=''
data_analyse['ind_terrain']=''
data_analyse['ind_teletrav']=''
data_analyse['ind_chomag']=''
data_analyse['ind_engagement']=''
data_analyse['ind_manag']=''

data_analyse['score_moral']=(((data_analyse['fatigue_num']+(data_analyse['stress_num_ind'])+
                               (data_analyse['sommeil_num_ind']))-4)/(12-4))*100

data_analyse['score_terrain']=(((data_analyse['soliterrain_num']+(5-data_analyse['secumax_num'])+
                                (5-data_analyse['secu_num'])+(5-data_analyse['clarteregle_num'])+
                                (data_analyse['application_num'])+(5-data_analyse['comprehension_num'])+
                                (5-data_analyse['intervention_num']))-4)/(28-4))*100

data_analyse['score_teletrav']=(((data_analyse['solitude_num']+(5-data_analyse['viepriv_num'])+
                                (5-data_analyse['maison_num'])+
                                (5-data_analyse['concentration_num']))-4)/(16-4))*100

data_analyse['score_chomag']=(((5-data_analyse['chomage_num']+
                                (5-data_analyse['relationcol_num']))-4)/(8-4))*100

data_analyse['score_engagement']=(((5-data_analyse['confent_num']+(5-data_analyse['adap_num'])+
                                (5-data_analyse['soutien_num'])+(5-data_analyse['soutcoll_num'])+
                                (5-data_analyse['perso_num']))-4)/(20-4))*100

data_analyse['score_manag']=(((5-data_analyse['infor_num']+(5-data_analyse['reu_num'])+
                                (5-data_analyse['echange_num'])+
                                (5-data_analyse['crise_num']))-4)/(16-4))*100

for i in range(0,len(list_ind_score)):
        for k in range(0,len(data_analyse)):
            if(data_analyse[list_ind_score[i]].iloc[k]<=25):
                data_analyse[list_ind[i]].iloc[k]=1
            if(25<data_analyse[list_ind_score[i]].iloc[k]<=50):
                data_analyse[list_ind[i]].iloc[k]=2    
            if(50<data_analyse[list_ind_score[i]].iloc[k]<=75):
                data_analyse[list_ind[i]].iloc[k]=3
            if(data_analyse[list_ind_score[i]].iloc[k]>75):
                data_analyse[list_ind[i]].iloc[k]=4
data_analyse
```

```python
list_var=[]
for i in range(0, len(lis_var)):
    k=lis_var[i]+'_num'
    list_var.append(k)

list_ind_glob =list_ind+list_var

#tri a plat sur les indicaure crées
for i in range(0,len(list_ind_glob)):
    tri=pd.crosstab(data_analyse[list_ind_glob[i]],margins=False,columns="count")
    tri=tri.reset_index()
    tri.columns=['mod','count']
    tri.dropna()
    tri=tri[tri['mod']!='']
    tri['prc']=tri['count']/tri['count'].sum()*100
    tri['var']=list_ind_glob[i]
    tri['mod'] = tri['mod'].astype(str)
    if ( i == 0):
        res_tri=tri
    else :
        res_tri=pd.concat([res_tri, tri])
        
res_tri.reset_index(drop=True, inplace=True)
ref['mod'] = ref['mod'].astype(str)
ref['id'] = ref['id'].astype(str)

for i in range(0,len(ref)):
    ref["mod"][i]=ref["mod"][i].replace('.0','')
for i in range(0,len(res_tri)):
    res_tri["mod"][i]=res_tri["mod"][i].replace('.0','')
    
res_tri=pd.merge(res_tri, ref, on=['var','mod'], how='outer')

# changer les valeurs manquante en '0'
for i in range (0,len(res_tri)):
    if(np.isnan(res_tri['count'].iloc[i])):
        res_tri['count'].iloc[i]=0
        res_tri['prc'].iloc[i]=0
res_tri['count'] = res_tri['count'].astype(int)
res_tri['socio'] = 'Ensemble'
res_tri['socio_mod'] = 'Ensemble'
resultat_tri=res_tri
resultat_tri
```

```python
res_tri_cr=pd.DataFrame()
for j in range(0,len(list_ind_glob)):
    for z in range(0,len(socio_var)):
        tri=pd.crosstab(data_analyse[list_ind_glob[j]],data_analyse[socio_var[z]])
        tri=tri.reset_index()
        for i in range(1,len(tri.columns)):
            tri1=[tri.iloc[:,0],tri.iloc[:,i]]
            tri1=pd.DataFrame(tri1)
            tri1=tri1.transpose()
            tri_columns=tri1.columns[1]    
            tri1.columns=['mod','count']
            tri1=tri1[tri1['mod']!='']
            tri1['prc']=0
            if (tri1['count'].sum()>0):
                tri1['prc']=tri1['count']/tri1['count'].sum()*100
            tri1['var']=list_ind_glob[j]
            tri1['mod'] = tri1['mod'].astype(str)
            tri1.reset_index(drop=True, inplace=True)
            for n in range(0,len(tri1)):
                tri1["mod"][n]=tri1["mod"][n].replace('.0','')
            #merge avec la table de référence :
            ref1=ref[ref['var']==list_ind_glob[j]]
            ref1.reset_index(drop=True, inplace=True)
            tri1=pd.merge(tri1, ref1, on=['var','mod'], how='outer')
            tri1['socio'] = socio_var[z]
            tri1['socio_mod'] = tri_columns
            tri1.reset_index(drop=True, inplace=True)
            
            res_tri_cr = pd.concat([res_tri_cr, tri1])
res_tri_cr.reset_index(drop=True, inplace=True)
res_tri_cr=res_tri_cr[res_tri_cr['socio_mod']!='']

for i in range (0,len(res_tri_cr)):
    if(np.isnan(res_tri_cr['count'].iloc[i])):
        res_tri_cr['count'].iloc[i]=0
        res_tri_cr['prc'].iloc[i]=0
        
resultat_tri=resultat_tri.append(res_tri_cr)
resultat_tri.reset_index(drop=True, inplace=True)
resultat_tri
```
